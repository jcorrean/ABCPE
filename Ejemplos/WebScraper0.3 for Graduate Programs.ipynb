```python
import requests
from bs4 import BeautifulSoup

url = "https://www.webometrics.info/en/Latin_America/Mexico"

# Send a GET request to the URL
response = requests.get(url)

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(response.content, 'html.parser')

# Find the table that contains the university information
table = soup.find('table', class_='sticky-enabled')

# Find all the rows in the table except the header row
rows = table.find_all('tr')[1:]

# Extract the university name and official URL for each row
for row in rows:
    columns = row.find_all('td')
    university_name = columns[1].text.strip()
    official_url_element = columns[1].find('a')
    if official_url_element:
        official_url = official_url_element['href']
        print(f"University: {university_name}")
        print(f"Official URL: {official_url}")
        print()
    else:
        print(f"University: {university_name}")
        print("Official URL not available")
        print()
```
